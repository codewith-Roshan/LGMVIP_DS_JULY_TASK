sepeating input column and the output column


```python
#sepearte features column and target column
data = df.values
x = data[:,0:4]
y = data[:,4]
```

spliting the data into traning and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

MODEL 1: Support vector machine algorithm

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train, y_train)
```

```python
# features = [[sepal_length, sepal_width, petal_length, petal_width]]
features = np.array([[5.1,	3.5,	1.4,	0.2,]]) #it is data of 0th row


model_prediction = model_svc.predict(features	)
print("model_prediction{}".format(model_prediction))
```

```python
from sklearn.metrics import accuracy_score
prediction1  = model_svc.predict(x_test)
print(accuracy_score(y_test,prediction1)*100)
```

MODEL 2: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
prediction2 = model_LR.predict(x_test)
#calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2)*100)
```

MODEL 3: Decision Tree Classifier


```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()
model_dtc.fit(x_train, y_train)
```

```python
prediction3 =model_dtc.predict(x_test)
#calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction3)*100)
```
